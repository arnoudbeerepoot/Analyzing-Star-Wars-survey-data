# Read in the data

```python
import pandas as pd
import matplotlib as plt
import numpy as np

star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head(10)
```

```python
# show the columms
star_wars.columns
```

A lot of 'Unnamed' columns are shown in the column list above

## Remove the rows where RespondentID is NaN

```python
# Number of rows in dataframe
row_count = len(star_wars)
row_count
```

```python
# Count the number of NaNs in RespondentID
Resp_Nans = star_wars["RespondentID"].isnull().sum().sum()
Resp_Nans
```

```python
# Remove the rows where RespondentID is NaN
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars.head(10)
```

```python
# Number of rows in dataframe
row_count = len(star_wars)
row_count
```

1186 is the correct number of rows which should be left in the star_wars dataframe

## Convert eight columns with Yes/No values to Boolean 

```python
# Convert Yes/No columns to Boolean
yes_no = {
    "Yes": True,
    "No": False
}
column1 = "Have you seen any of the 6 films in the Star Wars franchise?"
column2 = "Do you consider yourself to be a fan of the Star Wars film franchise?"

star_wars[column1] = star_wars[column1].map(yes_no)
star_wars[column2] = star_wars[column2].map(yes_no)

star_wars.head(10)
```

```python
# Change six columns which show the seen movies to Yes/No
import numpy as np #is used for the NaN value
movies={
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.NaN: False
}
columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6"
}
star_wars_columns = star_wars.columns[3:9]

for column in star_wars_columns:
    star_wars[column] = star_wars[column].map(movies)
star_wars.rename(columns=columns, inplace=True)
star_wars.head(5)
```

## Convert the six columns with ranking of the Star Wars movies from 1 (most favorite) to 6 (least favorite)

```python
# First each column will be converted to a float type
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
# Change the column names to a more descriptive name
cols = ["ranking_1", "ranking_2", "ranking_3", "ranking_4", "ranking_5", "ranking_6"]

ranking = {
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    "Unnamed: 10": "ranking_2",
    "Unnamed: 11": "ranking_3",
    "Unnamed: 12": "ranking_4",
    "Unnamed: 13": "ranking_5",
    "Unnamed: 14": "ranking_6"
}

star_wars.rename(columns=ranking, inplace=True)
star_wars[cols].head(5)
```

## Find the hightest-ranked movie

```python
# Calculate the mean value of the ranking columns
mean_ranking = star_wars[cols].mean()
mean_ranking
```

```python
# Create a Bar chart of each ranking
%matplotlib inline

mean_ranking.plot.bar()
```

From the bar plot above it can be seen that ranking_5 (Star Wars: Episode V The Empire Strikes Back) is the movie which is the most favorite (lowest mean value).

## Show the amount of people who have seen each movie

```python
# Calculate the sum of the seen columns
seen_columns = ["seen_1", "seen_2", "seen_3", "seen_4", "seen_5", "seen_6"]
seen_sum = star_wars[seen_columns].sum()
seen_sum.plot.bar()
```

In the plot above is shown that seen_5 (Star Wars: Episode V The Empire Strikes Back) is the movie which has the most viewers. When the two bar plots are compared you can see there is a correlation between the amount of viewers and ranking. More viewers leads to a higher ranking.

## Analyse the ranking between fans and non-fans

```python
# Split the data in fans and non-fans
fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"]== True]
non_fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"]== False]
```

### Ranking between fans and non-fans

```python
# cols is list with names of ranking columns
fans_mean = fans[cols].mean()
non_fans_mean = non_fans[cols].mean()

ind = np.arange(len(fans_mean))  # the x locations for the groups
width = 0.35  # the width of the bars

fig, ax = plt.pyplot.subplots()
fans1 = ax.bar(ind - width, fans_mean, width, color="Blue", label="fans")
fans2 = ax.bar(ind, non_fans_mean, width, color = "Red", label="non-fans")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Ranking')
ax.set_title('Ranking Star Wars movies between fans and non-fans')
ax.set_xticks(ind)
ax.set_xticklabels(('Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5', 'Rank6'))
ax.legend()
```

### Number of people saw the movies between fans and non-fans

```python
fans_seen = fans[seen_columns].sum()
non_fans_seen = non_fans[seen_columns].sum()

ind = np.arange(len(fans_seen))  # the x locations for the groups
width = 0.35  # the width of the bars

fig, ax = plt.pyplot.subplots()
fans1 = ax.bar(ind - width, fans_seen, width, color="Blue", label="fans")
fans2 = ax.bar(ind, non_fans_seen, width, color = "Red", label="non-fans")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Seen')
ax.set_title('Saw Star Wars movies between fans and non-fans')
ax.set_xticks(ind)
ax.set_xticklabels(('Seen1', 'Seen2', 'Seen3', 'Seen4', 'Seen5', 'Seen6'))
ax.legend()
```

In the plot above it can be seen that there are much more fans than non-fans who have ranked the movies. When it comes to ranking it can be seen that the fans ranking between good and less good movies has a bigger difference than the non-fans ranking. Fans give a high or a low ranking for each movie but non-fans give an average ranking for each movie.
